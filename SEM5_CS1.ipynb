<h2>Step 4: Splitting the data into training and testing data<h2>

```python
#Spliting the data into train (df_train) and test (df_test) data.
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(cancer, train_size = 0.70, test_size = 0.30, random_state = 100)
```

```python
#Using MinMaxScaler for scaling the cancer dataframe. 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
num_var = cancer.columns
```

```python
df_train[num_var] = scaler.fit_transform(df_train[num_var])
df_train.head()
```

```python
df_train.describe()
```

```python
#Assing our target variable to y_train and other independent variables to X_train.
y_train = df_train.pop('TARGET_deathRate')
X_train = df_train
```

```python
#Importing libraries for RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
#Using LinearRegeression fuction to fit X_train and y_train.
lgr = LinearRegression()
lgr.fit(X_train, y_train)
```

```python
#Using RFE to predict top 60 columns to be included for training the model.
rfe = RFE(lgr,60)
rfe = rfe.fit(X_train, y_train)
```

```python
#Viewing the list of columns with their support and ranking.
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
#Considering the columns which have RFE ranking as 1
col = X_train.columns[rfe.support_]
col
```

```python
X_train_rfe = X_train[col]
```

<h2>MODEL 1<h2>

```python
#Adding a constant using statsmodels.api
import statsmodels.api as sm
X_train_new = sm.add_constant(X_train_rfe)
```

```python
lgr = sm.OLS(y_train, X_train_new).fit()
```

```python
print(lgr.summary())
```

<h2>MODEL 2<h2>

```python
import statsmodels.api as sm
X_train_new1 = sm.add_constant(X_train_new)
```

```python
lgr = sm.OLS(y_train, X_train_new1).fit()
print(lgr.summary())
```

<h4>Dropping the columns having p-value greater than 0.05<h4>

```python
X_train_new1 = X_train_new.drop(['New Jersey','Mississippi','Michigan','Connecticut','PctEmployed16_Over'], axis = 1)
```

<h2>MODEL 3<h2>

```python
import statsmodels.api as sm
X_train_new2 = sm.add_constant(X_train_new1)
```

```python
lgr = sm.OLS(y_train, X_train_new2).fit()
print(lgr.summary())
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new1
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

<h4>VIF value less than 5 is acceptable so we will drop the columns having VIF value greater than 5<h4>

```python
#Dropping the columns with VIF > 5
X_train_new2 = X_train_new1.drop(['avgAnnCount','avgDeathsPerYear','PctWhite','PercentMarried','PctBlack','PctMarriedHouseholds'], axis = 1)
```

```python
#Removing the columns having p-vale >0.05
X_train_new3 = X_train_new2.drop(['Wisconsin','South Dakota','Oregon','Kansas'], axis = 1)
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new3
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
